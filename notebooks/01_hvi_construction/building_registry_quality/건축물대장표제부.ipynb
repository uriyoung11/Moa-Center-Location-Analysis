```python
import pandas as pd
cols = ['대지위치', '법정동코드명', '주용도코드명', '기타용도내용',
        '사용승인일자', '대지면적', '세대수', '주지번', '부지번']

df_원본 = pd.read_csv('../data/서울시 건축물대장 표제부.csv',
                      encoding='cp949',
                      encoding_errors='ignore',
                      on_bad_lines='skip',
                      usecols=cols)
```

```python
아파트_키워드 = ['아파트', '경비실', '지하주차장', '경로당', '주민공동시설', '관리사무소']

mask = (
    (df_원본['주용도코드명'] == '공동주택') &
    ~df_원본['기타용도내용'].str.contains('|'.join(아파트_키워드), na=False)
)

df = df_원본[mask].copy()

# 건축연도 추출 및 2022 이하 필터
df['건축연도'] = pd.to_numeric(
    df['사용승인일자'].astype(str).str[:4], errors='coerce')
df = df[df['건축연도'].notna() & (df['건축연도'] <= 2022)].copy()

print(f'최종 건물 수: {len(df):,}')
```

```python
'''
idx        # df의 원본 인덱스
lng        # 경도
lat        # 위도
hdong_code # 행정동 코드 (10자리, 예: 1150063000)
hdong_name # 행정동명 (예: 방화1동)
'''
import requests
import time
import os
import pandas as pd
from dotenv import load_dotenv

load_dotenv()
KAKAO_KEY = os.getenv('REST_API')
CHECKPOINT = '../data/checkpoint.csv'

def geocode(address):
    url = "https://dapi.kakao.com/v2/local/search/address.json"
    headers = {"Authorization": f"KakaoAK {KAKAO_KEY}"}
    params = {"query": address}
    try:
        res = requests.get(url, headers=headers, params=params, timeout=5)
        docs = res.json().get('documents', [])
        if docs:
            addr = docs[0].get('address', {})
            return {
                'lng': float(docs[0]['x']),
                'lat': float(docs[0]['y']),
                'hdong_code': addr.get('h_code'),
                'hdong_name': addr.get('region_3depth_h_name'),
                'b_code': addr.get('b_code')
            }
        return {'lng': None, 'lat': None, 'hdong_code': None, 'hdong_name': None, 'b_code': None}
    except:
        return {'lng': None, 'lat': None, 'hdong_code': None, 'hdong_name': None, 'b_code': None}

# 체크포인트 초기화
if os.path.exists(CHECKPOINT):
    os.remove(CHECKPOINT)

df_done = pd.DataFrame(columns=['idx', 'lng', 'lat', 'hdong_code', 'hdong_name', 'b_code'])
results = []
total = len(df)

for i, (idx, row) in enumerate(df.iterrows()):
    result = geocode(row['대지위치'])
    result['idx'] = idx
    results.append(result)

    if len(results) % 1000 == 0:
        df_done = pd.concat([df_done, pd.DataFrame(results)], ignore_index=True)
        df_done.to_csv(CHECKPOINT, index=False, encoding='utf-8-sig')
        results = []
        print(f'{len(df_done):,} / {total:,} 완료')

    time.sleep(0.05)

if results:
    df_done = pd.concat([df_done, pd.DataFrame(results)], ignore_index=True)
    df_done.to_csv(CHECKPOINT, index=False, encoding='utf-8-sig')

print(f'완료. 총 {len(df_done):,}건')
```

```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point

# 1. shapefile 로드
gdf_hdong = gpd.read_file('../data/hdong_data/bnd_dong_11_2025_2Q/bnd_dong_11_2025_2Q.shp')
gdf_hdong = gdf_hdong.to_crs(epsg=4326)
gdf_hdong = gdf_hdong.rename(columns={'ADM_CD': 'hdong_code', 'ADM_NM': 'hdong_name'})

# 2. checkpoint 로드 후 좌표 있는 것만
df_geo = pd.read_csv('../data/checkpoint.csv')
df_valid = df_geo[df_geo['lng'].notna() & df_geo['lat'].notna()].copy()

# 3. GeoDataFrame 변환
gdf_bldg = gpd.GeoDataFrame(
    df_valid,
    geometry=gpd.points_from_xy(df_valid['lng'], df_valid['lat']),
    crs='EPSG:4326'
)

# 4. sjoin
gdf_joined = gpd.sjoin(
    gdf_bldg,
    gdf_hdong[['hdong_code', 'hdong_name', 'geometry']],
    how='left',
    predicate='within'
)

# 5. 통폐합 규칙 적용
dong_rename = {'상일1동': '상일동', '상일2동': '강일동'}
code_rename = {'11250760': '11250760', '11250770': '11250750'}

gdf_joined['hdong_name_right'] = gdf_joined['hdong_name_right'].replace(dong_rename)
gdf_joined['hdong_code_right'] = gdf_joined['hdong_code_right'].astype(str).replace(code_rename)


# 6. 원본 df와 merge (좌표 있는 것만)
df_geo_valid = df_geo[df_geo['lng'].notna() & df_geo['lat'].notna()].set_index('idx')
df_final = df.loc[df.index.isin(df_geo_valid.index)].join(df_geo_valid[['lng', 'lat', 'b_code']])
df_final['hdong_code'] = gdf_joined.set_index('idx')['hdong_code_right']
df_final['hdong_name'] = gdf_joined.set_index('idx')['hdong_name_right']

# 7. 컬럼명 정리 및 PNU 생성
df_final = df_final.rename(columns={
    '사용승인일자': 'SYDATE',
    '대지면적':    'total_area',
    '세대수':      'total_units'
})

df_final['b_code']     = df_final['b_code'].astype(str).str.split('.').str[0].str.zfill(10)
df_final['bun_padded'] = df_final['주지번'].astype(str).str.split('.').str[0].str.zfill(4)
df_final['ji_padded']  = df_final['부지번'].astype(str).str.split('.').str[0].str.zfill(4)
df_final['PNU']        = df_final['b_code'] + '0' + df_final['bun_padded'] + df_final['ji_padded']
# 구 이름 추출
df_final['gu_name'] = df_final['대지위치'].str.extract(r'서울특별시\s+(\S+구)')

# 8. 최종 저장
cols = ['대지위치', 'gu_name', 'hdong_code', 'hdong_name',
        'SYDATE', 'total_area', 'total_units',
        'PNU', 'lng', 'lat']

df_final[cols].to_csv('../data/건축물대장_빅캠반입용.csv',
                      index=False, encoding='utf-8-sig')

print(f'반입용 데이터: {df_final[cols].shape}')
print(df_final[cols].isnull().sum())
print(f'\n총 행정동 수: {df_final["hdong_code"].nunique()}')
```

```python
# 노후도용: hdong_code 있는 것 (SYDATE 결측 0이니까 그대로)
old_count = gdf_joined[gdf_joined['hdong_code_right'].notna()].groupby(
    ['hdong_code_right', 'hdong_name_right']).size().reset_index(name='노후도_건물수')

# 집적도용: hdong_code 있고 total_area > 0 이고 total_units notna
df_final_valid = df_final[
    df_final['hdong_code'].notna() &
    (df_final['total_area'] > 0) &
    df_final['total_area'].notna() &
    df_final['total_units'].notna()
]
density_count = df_final_valid.groupby(['hdong_code', 'hdong_name']).size().reset_index(name='집적도_건물수')

# 합쳐서 비교
dong_count_final = pd.merge(
    old_count,
    density_count,
    left_on='hdong_code_right',
    right_on='hdong_code',
    how='outer'
)

dong_count_final = dong_count_final.sort_values('노후도_건물수', ascending=False)
print(f'총 행정동 수: {len(dong_count_final)}')
print(dong_count_final[['hdong_code_right', 'hdong_name_right', '노후도_건물수', '집적도_건물수']].to_string())
```
