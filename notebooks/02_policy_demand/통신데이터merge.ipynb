```python
df = df_all_years.copy()
df.columns = df.columns.str.strip()

# 자치구 컬럼명 자동 탐지
gu_col = None
for col in df.columns:
    if '자치구' in col or col in ['구', 'gu', 'gu_nm']:
        gu_col = col
        break

if gu_col is None:
    raise ValueError(f"자치구 컬럼을 찾을 수 없습니다. 컬럼 목록: {df.columns.tolist()}")

print(f"✅ 자치구 컬럼 탐지: '{gu_col}'")

# 고유 키 생성 (자치구 + 행정동)
df['_동키'] = df[gu_col] + '_' + df['행정동']

# 고령층 필터링
df_elderly = df[df['연령대'] >= 65].copy()

# ============================================================
# STEP 2: 컬럼 탐지
# ============================================================
column_mapping = {}
for col in df.columns:
    if '통화' in col and '대상자' in col:
        column_mapping['통화대상자'] = col
    elif '문자' in col and '대상자' in col:
        column_mapping['문자대상자'] = col
    elif '체류시간' in col and '평일' in col:
        column_mapping['평일_체류시간'] = col

# ============================================================
# STEP 3: 3개 지수 생성
# ============================================================

# --- A. 사회적 고립도 ---
통화 = column_mapping.get('통화대상자')
문자 = column_mapping.get('문자대상자')
if 통화 and 문자:
    df_elderly['_소통합계'] = df_elderly[통화] + df_elderly[문자]

# --- B. 재택기반 고립 ---
평일체류 = column_mapping.get('평일_체류시간')
if 평일체류:
    df_elderly['_평일재택률'] = (df_elderly[평일체류] / 1440).clip(0, 1)

# --- C. 1인가구 지수 (고령 가중치 추가) ---
if '1인가구수' in df.columns and '총인구수' in df.columns:
    # 동키 기준으로 고령 비율 계산 (자치구 구분 포함)
    dong_elderly_ratio = (
        df[df['연령대'] >= 65]
        .groupby('_동키')['총인구수'].sum()
        /
        df.groupby('_동키')['총인구수'].sum()
    ).rename('고령비율').reset_index()

    df_elderly = df_elderly.merge(dong_elderly_ratio, on='_동키', how='left')
    df_elderly['고령비율'] = df_elderly['고령비율'].fillna(0)

    df_elderly['_1인가구_기본'] = df_elderly['1인가구수'] / (df_elderly['총인구수'] + 1)
    df_elderly['_1인가구_가중'] = df_elderly['_1인가구_기본'] * (1 + 0.5 * df_elderly['고령비율'])

# ============================================================
# STEP 4: 동키 기준으로 집계
# ============================================================
agg_cols = {}
if '_소통합계' in df_elderly.columns:
    agg_cols['_소통합계'] = 'mean'
if '_평일재택률' in df_elderly.columns:
    agg_cols['_평일재택률'] = 'mean'
if '_1인가구_가중' in df_elderly.columns:
    agg_cols['_1인가구_가중'] = 'mean'

# 자치구·행정동 이름도 함께 보존
agg_cols[gu_col] = 'first'
agg_cols['행정동'] = 'first'

df_dong = df_elderly.groupby('_동키').agg(agg_cols).reset_index()

# 중복 동 이름 확인 출력
dup_check = df_dong.groupby('행정동')['_동키'].count()
dup_dongs = dup_check[dup_check > 1].index.tolist()
if dup_dongs:
    print(f"⚠️ 동일 이름 행정동 (자치구 구분 처리됨): {dup_dongs}")

# ============================================================
# STEP 5: MinMaxScaler → 0~100 스케일링
# ============================================================
scaler = MinMaxScaler(feature_range=(0, 100))

if '_소통합계' in df_dong.columns:
    df_dong['사회적_고립도'] = (
        100 - scaler.fit_transform(df_dong[['_소통합계']]).flatten()
    ).clip(0, 100)  # 부동소수점 오차 제거

if '_평일재택률' in df_dong.columns:
    df_dong['재택기반_고립'] = (
        scaler.fit_transform(df_dong[['_평일재택률']]).flatten()
    ).clip(0, 100)

if '_1인가구_가중' in df_dong.columns:
    df_dong['1인가구_지수'] = (
        scaler.fit_transform(df_dong[['_1인가구_가중']]).flatten()
    ).clip(0, 100)

# ============================================================
# STEP 6: 최종 취약 생활지수
# ============================================================
weight_map = {
    '사회적_고립도': 0.40,
    '재택기반_고립': 0.30,
    '1인가구_지수':  0.30,
}

available = [c for c in weight_map if c in df_dong.columns]
total_w = sum(weight_map[c] for c in available)

df_dong['취약_생활지수'] = sum(
    df_dong[c] * (weight_map[c] / total_w) for c in available
).clip(0, 100)

# ============================================================
# STEP 7: 자치구 / 행정동 분리 컬럼으로 출력
# ============================================================
# _동키가 "강남구_신사동" 형태이므로 _ 기준으로 분리
df_dong[['자치구', '행정동']] = df_dong['_동키'].str.split('_', n=1, expand=True)

print(df_dong[['자치구', '행정동', '사회적_고립도', '재택기반_고립', '1인가구_지수', '취약_생활지수']]
      .sort_values('취약_생활지수', ascending=False)
      .head(20)
      .to_string(index=False))

# 신사동 확인
print("\n📍 신사동 확인:")
print(df_dong[df_dong['행정동'].str.contains('신사')][
    ['자치구', '행정동', '사회적_고립도', '재택기반_고립', '1인가구_지수', '취약_생활지수']
].to_string(index=False))
```
